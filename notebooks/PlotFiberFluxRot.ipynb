```python
from pfs.utils.fiberids import FiberIds
fiberIds = FiberIds()
```

```python
def plot(axs, flux_rot):
    x_array=[]
    y_array=[]
    for fid in flux_rot[0]:
        x,y = fiberIds.xyForCobras(int(fid))
        x_array.append(x)
        y_array.append(y)
    axs.scatter(x_array, y_array, c=flux_rot[1])
```

```python
flux_rot_0=np.load('/tigress/ncaplar/PIPE2D-902/flux_rot_0.npy')
flux_rot_ratio_0_p90=np.load('/tigress/ncaplar/PIPE2D-902/flux_rot_ratio_0_p90.npy')
flux_rot_ratio_0_m90=np.load('/tigress/ncaplar/PIPE2D-902/flux_rot_ratio_0_m90.npy')
flux_rot_ratio_0_180=np.load('/tigress/ncaplar/PIPE2D-902/flux_rot_ratio_0_180.npy')
```

```python
fig, axs = plt.subplots(2, 2)
plot(axs[0, 0], flux_rot_0)
axs[0, 0].set_title('flux_rot_0')
plot(axs[0, 1], flux_rot_ratio_0_p90)
axs[0, 1].set_title('flux_rot_ratio_0_p90')
plot(axs[1, 0], flux_rot_ratio_0_m90)
axs[1, 0].set_title('flux_rot_ratio_0_m90')
plot(axs[1, 1], flux_rot_ratio_0_180)
axs[1, 1].set_title('flux_rot_ratio_0_180')
```
